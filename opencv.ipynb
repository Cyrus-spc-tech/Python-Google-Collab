```python
import cv2 

image1 = cv2.imread("16.jpg")
image2 = cv2.imread("13.jpg")

if image1.shape == image2.shape:
    add = cv2.add(image1, image2)
    cv2.imshow('Weighted Sum', add)
    cv2.waitKey(0)
    cv2.destroyAllWindows()
    
else:
    print("Images do not have the same shape and cannot be added.")
```
